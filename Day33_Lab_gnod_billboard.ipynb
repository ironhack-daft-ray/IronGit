```python
# 1. import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.billboard.com/charts/hot-100"
```

```python
text = requests.get(url).text
```

```python
# parse the element
soup = BeautifulSoup(text, parser='html5lib')
```

```python
soup.head()
```

# SONG NAMES SCRAPPING

```python
# Retrieve all the song names
soup.find_all("span", class_="chart-element__information__song text--truncate color--primary")
```

```python
all_songs = []
for i in soup.find_all("span", class_="chart-element__information__song text--truncate color--primary"):
    all_songs.append(i.get_text())
```

```python
all_songs
```

# ARTIST SCRAPPING

```python
# Retrieve all the artist names
soup.find_all("span", class_="chart-element__information__artist text--truncate color--secondary")
```

```python
all_artists = []
for i in soup.find_all("span", class_="chart-element__information__artist text--truncate color--secondary"):
    all_artists.append(i.get_text())
```

```python
all_artists
```

# include in a dataframe

```python
billboard = pd.DataFrame({"songs":all_songs, "artist": all_artists})
```

```python
billboard.describe()
```

# Adding an extra source (Spotify charts)

```python
url2 = 'https://spotifycharts.com/regional/global/weekly/latest'
```

```python
text2 = requests.get(url2).text
```

```python
soup2 = BeautifulSoup(text2, parser='html5lib')
soup2
```

## spotify song names

```python
#soup2.find_all("td", class_="chart-table-track")
```

```python
# Retrieve all the song names
soup2.find_all("strong")
```

```python
all_songs2 = []
#for i in soup2.find_all("td", class_="chart-table-track"):
for i in soup2.find_all("strong"):
    all_songs2.append(i.get_text().replace("\n\n", "").strip())
all_songs2
```

```python
len(all_songs2)
```

## spotify artist

```python
#retrieve artist name
all_artists2 = []
for i in soup2.select('td.chart-table-track > span'):
    all_artists2.append(i.get_text().replace("by ", "").replace("\n\n", "").strip())
all_artists2
```

```python
len(all_artists2)
```

## include in dataframe spotify

```python
billboard2 = pd.DataFrame({"songs":all_songs2, "artist": all_artists2})
```

```python
billboard2.head()
```

```python
billboard2.shape
```

## merging the data frames

```python
billboard_final = pd.concat([billboard, billboard2])
billboard_final.shape
```
